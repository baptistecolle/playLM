```python
from llama_cpp import Llama

llm = Llama(model_path="../model/wizardLM-7B.ggmlv3.q4_1.bin", logits_all=True)
```

```python
prompt = f"""
You are a video game player. \
This is current state of board you are in position 0. \
You can use the following actions: 'left', 'down', 'right', 'up'. \
Please output one of the possible actions? \
If you are unsure, please output an action as you are learning how to use the controllers of this video game. \
ACTION SELECTED IS:"""
output = llm(prompt, max_tokens=32, stop=["Q:", "\n"], echo=True, logprobs=10)
print(output)
```
